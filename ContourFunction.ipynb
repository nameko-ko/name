```python
#sinθ型のContourFunction。棄却
import math
import numpy as np
import matplotlib.pyplot as plt 

a = 1
b = 1
c = 1
d = 0
m = 1

r = []
theta = np.linspace(0, 2*np.pi, 1000)
for i in range(len(theta)):
    r.append(a*(2.0 - b * abs(math.sin(1.5 * c *(theta[i]-d))**m)))

x = r * np.cos(theta)
y = r * np.sin(theta)

fig = plt.figure(figsize = (10,6))
#グラフを描画するsubplot領域を作成。
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
#フォント設定
plt.rcParams["font.family"] = "Arial"
ax1.plot(theta, r)
ax2.plot(x, y)
ax1.set_xlabel('θ')
ax2.set_xlabel('x')
ax1.set_ylabel('r')
ax2.set_ylabel('y')
#テキストボックスの生成
textstr = '\n'.join((
    r'$a=%.2f$' % (a, ),
    r'$b=%.2f$' % (b, ),
    r'$c=%.2f$' % (c, ),
    r'$d=%.2f$' % (d, ),
    r'$m=%.2f$' % (m, )))
# 数式の文字列を生成
equation_str = r'$r=a×(2-b|sin(1.5c(θ - d))|^m)$'
# テキストボックスの属性を設定
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# テキストボックスを座標 (0.05, 0.95) に配置
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
ax1.text(0.05, 0.95, equation_str, transform=ax1.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)
ax1.set_xlim([0, 6.3])
ax1.set_ylim([0, 4])
ax2.set_xlim([-5, 5])
ax2.set_ylim([-3, 3])

plt.show()
```

```python
#secθ型の一つの条件のみ表示する
import math
import numpy as np
import matplotlib.pyplot as plt 

a = 1 #拡大
b = 2 #遷移1
c = 0 #回転
d = 1 #遷移2
cons = np.pi/3 #定数


r = []
theta = np.linspace(0, 2*np.pi, 1000)
for i in range(len(theta)):
    tmp = (d*theta[i]-c) % (np.pi/1.5)
    #r.append(a*(2.0 - b * abs(math.sin(1.5 * c *(theta[i]-d))**m)))
    r.append(a*(b*(1/math.cos(tmp-cons)-2)+2))

x = r * np.cos(theta)
y = r * np.sin(theta)

fig = plt.figure(figsize = (10,6))
#グラフを描画するsubplot領域を作成。
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
#フォント設定
plt.rcParams["font.family"] = "Arial"
ax1.plot(theta, r)
ax2.plot(x, y)
ax1.set_xlabel('θ')
ax2.set_xlabel('x')
ax1.set_ylabel('r')
ax2.set_ylabel('y')
#テキストボックスの生成
textstr = '\n'.join((
    r'$a=%.2f$' % (a, ),
    r'$b=%.2f$' % (b, ),
    r'$c=%.2f$' % (c, ),
    r'$d=%.2f$' % (d, ),))
# 数式の文字列を生成
equation_str = r'$r=a(b(sec(θ - π/3)-2)+2)$'
# テキストボックスの属性を設定
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# テキストボックスを座標 (0.05, 0.95) に配置
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
ax1.text(0.05, 0.95, equation_str, transform=ax1.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)
ax1.set_xlim([0, 6.3])
ax1.set_ylim([0, 4])
ax2.set_xlim([-5, 5])
ax2.set_ylim([-3, 3])
#余白の自動調整
plt.tight_layout()

path = "figure_b_{}_d_{}".format(b,d)
plt.savefig("/Users/nametakouhei/name/画像ファイル/{}".format(path),bbox_inches='tight')
plt.show()
```

```python
#マトリックス画像保存
import math
import numpy as np
import matplotlib.pyplot as plt 

a = 1  # 拡大
b = np.arange(0, 2.0, 0.25)  # 遷移1
c = 0  # 回転
d = np.arange(0, 5/3, 1/3)  # 遷移2
cons = np.pi/3  # 定数

theta = np.linspace(0, 2*np.pi, 200)

# グラフのサイズ
num_rows = min(len(b), len(d))
num_cols = max(len(b), len(d))

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 15), sharex=True, sharey=True)

for j, bj in enumerate(b):
    for i, di in enumerate(d):
        if i < num_rows and j < num_cols:  # 現在のインデックスが有効な範囲内か確認
            r = []
            for k in range(len(theta)):
                tmp = (di*theta[k]-c) % (np.pi/1.5)
                r.append(a*(bj*(1/math.cos(tmp-cons)-2)+2))

            x = np.array(r) * np.cos(theta)
            y = np.array(r) * np.sin(theta)

            ax1 = axs[i, j]
            #ax2 = ax1.inset_axes([0.5, 0.5, 0.45, 0.45])

            textstr = '\n'.join((
                r'$b=%.2f$' % (bj, ),
                r'$d=%.2f$' % (di, ),))

            #equation_str = r'$r=a(b(\sec(θ - π/3)-2)+2)$'
            props = dict(boxstyle='round', facecolor='white', alpha=0.5)

            ax1.plot(x, y)
            ax1.set_xlabel('x')
            ax1.set_ylabel('y')
            ax1.set_xlim([-3, 3])
            ax1.set_ylim([-4, 4])

            ax1.text(0.05, 0.95, textstr, transform=ax1.transAxes, fontsize=10,
                     verticalalignment='top', bbox=props)

# レイアウトの調整
fig.tight_layout()
plt.subplots_adjust(wspace=-0.012,hspace=0)

# 保存
plt.savefig("/Users/nametakouhei/name/画像ファイル/output_matrix.png")

# 表示
plt.show()
```
