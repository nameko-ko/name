```python
#sinθ型のContourFunction。棄却
import math
import numpy as np
import matplotlib.pyplot as plt 

a = 1
b = 1
c = 1
d = 0
m = 1

r = []
theta = np.linspace(0, 2*np.pi, 1000)
for i in range(len(theta)):
    r.append(a*(2.0 - b * abs(math.sin(1.5 * c *(theta[i]-d))**m)))

x = r * np.cos(theta)
y = r * np.sin(theta)

fig = plt.figure(figsize = (10,6))
#グラフを描画するsubplot領域を作成。
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
#フォント設定
plt.rcParams["font.family"] = "Arial"
ax1.plot(theta, r)
ax2.plot(x, y)
ax1.set_xlabel('θ')
ax2.set_xlabel('x')
ax1.set_ylabel('r')
ax2.set_ylabel('y')
#テキストボックスの生成
textstr = '\n'.join((
    r'$a=%.2f$' % (a, ),
    r'$b=%.2f$' % (b, ),
    r'$c=%.2f$' % (c, ),
    r'$d=%.2f$' % (d, ),
    r'$m=%.2f$' % (m, )))
# 数式の文字列を生成
equation_str = r'$r=a×(2-b|sin(1.5c(θ - d))|^m)$'
# テキストボックスの属性を設定
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# テキストボックスを座標 (0.05, 0.95) に配置
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
ax1.text(0.05, 0.95, equation_str, transform=ax1.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)
ax1.set_xlim([0, 6.3])
ax1.set_ylim([0, 4])
ax2.set_xlim([-5, 5])
ax2.set_ylim([-3, 3])

plt.show()
```

```python
#secθ型のContourFunction。採用。
import math
import numpy as np
import matplotlib.pyplot as plt 

a = 1 #拡大
b = 0 #遷移1
c = 0 #回転
d = 1 #遷移2
cons = np.pi/3 #定数


r = []
theta = np.linspace(0, 2*np.pi, 1000)
for i in range(len(theta)):
    tmp = (d*theta[i]-c) % (np.pi/1.5)
    #r.append(a*(2.0 - b * abs(math.sin(1.5 * c *(theta[i]-d))**m)))
    r.append(a*(b*(1/math.cos(tmp-cons)-2)+2))

x = r * np.cos(theta)
y = r * np.sin(theta)

fig = plt.figure(figsize = (10,6))
#グラフを描画するsubplot領域を作成。
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
#フォント設定
plt.rcParams["font.family"] = "Arial"
ax1.plot(theta, r)
ax2.plot(x, y)
ax1.set_xlabel('θ')
ax2.set_xlabel('x')
ax1.set_ylabel('r')
ax2.set_ylabel('y')
#テキストボックスの生成
textstr = '\n'.join((
    r'$a=%.2f$' % (a, ),
    r'$b=%.2f$' % (b, ),
    r'$c=%.2f$' % (c, ),
    r'$d=%.2f$' % (d, ),))
# 数式の文字列を生成
equation_str = r'$r=a(b(sec(θ - π/3)-2)+2)$'
# テキストボックスの属性を設定
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

# テキストボックスを座標 (0.05, 0.95) に配置
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
ax1.text(0.05, 0.95, equation_str, transform=ax1.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)
ax1.set_xlim([0, 6.3])
ax1.set_ylim([0, 4])
ax2.set_xlim([-5, 5])
ax2.set_ylim([-3, 3])
#余白の自動調整
plt.tight_layout()

path = "figure_b_{}_d_{}".format(b,d)
plt.savefig("/Users/nametakouhei/name/画像ファイル/{}".format(path),bbox_inches='tight')
plt.show()
```
