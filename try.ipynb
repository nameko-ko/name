```python
import cv2
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
import pixel_converter
from assemble_frames_with_order_numbers import assemble_frames_with_order_numbers
from table_image_generator import create_table_image, find_min_max_avg_table

def SearchCF(a,b,c,d,theta_cal, r_obs):
    r_cal = []
    for i in range(len(theta_cal)):
        tmp = (d/3*theta_cal[i]-c) % (np.pi/1.5)
        r_cal.append(a*(b*(1/math.cos(tmp-cons)-2)+2))
    MSE = 0
    for r_obs, r_cal in zip(r_obs, r_cal):
        MSE += (r_obs - r_cal) ** 2 /r_cal**2
    MSE /= len(r_obs)
    return MSE

# 画像ファイルのパス
image_path = "/Users/nametakouhei/name/画像ファイル/triangle.bmp"

# 画像の読み込み
color_image = cv2.imread(image_path, cv2.IMREAD_COLOR)
if color_image is None:
    print(f"Error: Failed to load image from {image_path}")
    sys.exit(1)

# 画像処理
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
blurred_image = cv2.bilateralFilter(gray_image, 7, 15, 15)
denoised_image = cv2.medianBlur(blurred_image, 5)#追加項目
ret, binary = cv2.threshold(denoised_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# 輪郭の抽出
contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

# 輪郭の条件設定
conditions = {
    "min_contour_area": 2000,
    "max_contour_area": 5000,
}

filtered_contours = []
area_data = []
perimeter_data = []
moment_data = []
max_radius = 0

for idx, contour in enumerate(contours):
    contour_area = cv2.contourArea(contour)

    if (
        ("min_contour_area" not in conditions or contour_area >= conditions["min_contour_area"]) and
        ("max_contour_area" not in conditions or contour_area <= conditions["max_contour_area"])
    ):
        filtered_contours.append(contour)
        area_data.append(contour_area)
        perimeter_data.append(cv2.arcLength(contour, True))
        M = cv2.moments(contour)
        filtered_cx = int(M['m10'] / M['m00']) if M["m00"] != 0 else 0
        filtered_cy = int(M['m01'] / M['m00']) if M["m00"] != 0 else 0
        moment_data.append((filtered_cx, filtered_cy))

        # 輪郭上の点と重心の距離を計算し、最大距離を半径とし直径を求める
        for point in contour:
            x, y = point[0]
            distance = math.sqrt((x - filtered_cx)**2 + (y - filtered_cy)**2)
            if distance > max_radius:
                max_radius = distance * 2

# 処理結果の表示（例として最初のフレームのみ）
result_image = color_image.copy()
for contour, moment in zip(filtered_contours, moment_data):
    x_coordinates = [point[0][0] for point in contour]
    y_coordinates = [point[0][1] for point in contour]

    min_x = min(x_coordinates)
    max_x = max(x_coordinates)
    min_y = min(y_coordinates)
    max_y = max(y_coordinates)
    base_length = max_x - min_x
    height = max_y - min_y

    M = cv2.moments(contour)
    rel_con = []
    if M["m00"] != 0:
        filtered_cx, filtered_cy = moment
        area = cv2.contourArea(contour)
        # 重心に点を描画
        cv2.circle(result_image, (filtered_cx, filtered_cy), 5, (0, 0, 255), -1)
        # 輪郭を描画
        cv2.drawContours(result_image, [contour], -1, (0, 255, 0), 2)
        rel_con = contour-moment
    #抽出したcontourの相対座標をx,yリストに分離
    con_x = []
    con_y = []
    for i,con in enumerate(rel_con):
        con_x.append(con[0][0])
        con_y.append(con[0][1]) 
    #r,θの計算
    r_cal = []
    r_obs = []
    theta_cal = []
    theta_obs = []
    for i,_ in enumerate(con_x):
        r_obs.append(np.sqrt(con_x[i]**2 + con_y[i]**2))
        theta_obs.append(np.arctan2(con_y[i],con_x[i]))
    theta_cal = theta_obs
    #最適なa,b,c,dの探索。def SearchCF
    a = [[i * 1 for i in range(int(30 / 1))]]
    b = [i * 0.1 for i in range(int(1.81 / 0.1))]#[0, 0.1, ...1.7]のようなリスト
    c = [i * 0.1 for i in range(int(2.01 / 0.1))]#[0, 0.1, ...2.0]のようなリスト
    d = [0,1,2,3,4,5]
    cons = np.pi/3 #定数
    ans_a = 0
    ans_b = 0
    ans_c = 0
    ans_d = 0
    S = 100
    MSE = 0
    for a_val in a:
        for b_val in b:
            for c_val in c:
                for d_val in d:
                    a_val = max_radius / (1+b_val)
                    MSE_tmp = SearchCF(a_val, b_val, c_val, d_val, theta_cal, r_obs)
                    if MSE_tmp < MSE:
                        MSE = MSE_tmp
                        ans_a = a_val
                        ans_b = b_val
                        ans_c = c_val
                        ans_d = d_val

    print("a :" + str(ans_a))
    print("b :" + str(ans_b))
    print("c :" + str(ans_c))
    print("d :" + str(ans_d))


# 結果の表示
plt.imshow(result_image)
plt.title('Processed Image')
plt.savefig("/Users/nametakouhei/name/画像ファイル/Processed Image.png")
plt.show()
```
